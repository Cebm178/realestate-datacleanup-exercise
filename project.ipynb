#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)

```python
# TODO

fuen_sub_filtered = ds_filtered_2[ds_filtered_2['level5'] == 'Fuenlabrada']
laga_sub_filtered = ds_filtered_2[ds_filtered_2['level5'] == 'Leganés']
geta_sub_filtered = ds_filtered_2[ds_filtered_2['level5'] == 'Getafe']
alco_sub_filtered = ds_filtered_2[ds_filtered_2['level5'] == 'Alcorcón']

locations = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
subsets = {loc: ds_filtered_2[ds_filtered_2['level5'] == loc] for loc in locations}

for loc, subset in subsets.items():
    plt.figure()  # Create a new figure for each plot
    plt.scatter(subset['surface'], subset['price'], alpha=0.5, label=loc)
    
    # Customize the plot
    plt.xlabel('Surface (m2)')
    plt.ylabel('Price ($)')
    plt.title(f'Scatter Plot of Price vs. Surface for {loc}')
    plt.legend(loc='upper left')
    plt.grid(True)

    plt.show()

```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python

south_belt_populations = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
south_belt_ds_filtered_2 = ds_filtered_2[ds_filtered_2['population'].isin(south_belt_populations)]

color_map = {
    'Fuenlabrada': 'Orange',
    'Leganés': 'Blue',
    'Getafe': 'Red',
    'Alcorcón': 'Purple'
}

plt.figure(figsize=(10, 8))

for population, color in color_map.items():
    subset = south_belt_ds_filtered_2[south_belt_ds_filtered_2['population'] == population]
    plt.scatter(subset['longitude'], subset['latitude'], c=color, label=population, alpha=0.5)

# Customize the plot
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Coordinates of the South Belt of Madrid by Population')
plt.legend()
plt.grid(True)
plt.show()
```
